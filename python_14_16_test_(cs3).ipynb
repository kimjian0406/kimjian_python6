```python
# 초기 재고 설정
stock = {
    '팥붕어빵': {"가격": 1000, "재고": 10, "총 판매개수": 0},
    '슈크림붕어빵': {"가격": 1200, "재고": 8, "총 판매개수": 0},
    '초코붕어빵': {"가격": 1500, "재고": 5, "총 판매개수": 0},
}

def display_stock():
    """현재 재고를 출력하는 함수"""
    print("\n[현재 재고 현황]")
    for name, info in stock.items():
        print(f"- {name}: {info['재고']}개 (가격: {info['가격']}원)")

def 관리자모드():
    """관리자 모드"""
    while True:
        display_stock()
        add_boong = input("\n추가할 붕어빵 이름을 입력하거나 '종료'를 입력하세요: ").strip()
        if add_boong == "종료":
            print("재고 추가를 종료합니다.")
            break

        if add_boong not in stock:
            print(f"'{add_boong}'은(는) 존재하지 않습니다. 다시 입력하세요.")
            continue

        try:
            how_boong = int(input("추가할 개수를 입력하세요: "))
            stock[add_boong]["재고"] += how_boong
            print(f"'{add_boong}' 재고가 {how_boong}개 추가되었습니다!")
        except ValueError:
            print("잘못된 입력입니다. 숫자로 입력해주세요!")

def 판매모드():
    """판매 모드"""
    receipt = {}  # 주문 내역 저장
    total_price = 0  # 총 금액 계산
    while True:
        display_stock()
        order = input("\n주문할 붕어빵 이름을 입력하거나 '종료'를 입력하세요: ").strip()
        if order == "종료":
            print("\n[영수증]")
            for name, quantity in receipt.items():
                print(f"- {name}: {quantity}개 (가격: {stock[name]['가격']}원)")
            print(f"총 금액: {total_price}원")
            print("판매를 종료합니다.")
            break

        if order not in stock:
            print(f"'{order}'은(는) 존재하지 않습니다. 다시 입력하세요.")
            continue

        try:
            order_q = int(input(f"'{order}'을(를) 몇 개 주문하시겠습니까? "))
            if stock[order]["재고"] >= order_q:
                stock[order]["재고"] -= order_q
                stock[order]["총 판매개수"] += order_q
                total_price += stock[order]["가격"] * order_q
                receipt[order] = receipt.get(order, 0) + order_q
                print(f"'{order}' {order_q}개가 추가되었습니다! 현재 총액: {total_price}원")
            else:
                print(f"재고가 부족합니다. 현재 재고: {stock[order]['재고']}개")
        except ValueError:
            print("잘못된 입력입니다. 숫자로 입력해주세요!")

while True:
    try:
        print("\n[붕어빵 시스템]")
        id = input("실행할 모드를 선택하세요: 1. 관리, 2. 판매, 3. 종료: ").strip()

        if id in ["1", "관리"]:
            관리자모드()
        elif id in ["2", "판매"]:
            판매모드()
        elif id in ["3", "종료"]:
            print("\n시스템을 종료합니다. 감사합니다!")
            break
        else:
            print("잘못된 선택입니다. 다시 입력해주세요.")

    except Exception as e:
        print(f"예기치 못한 오류가 발생했습니다: {e}")
```
